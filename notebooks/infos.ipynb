The Two Problems & Solutions
The Java Missing Link:

The Problem: Spark runs on a Java Virtual Machine (JVM). Even though Java was installed, Spark on Windows specifically looks for an environment variable called JAVA_HOME. Without it, Spark instantly crashes.

The Solution: We manually injected the JAVA_HOME variable into your script, pointing exactly to your Eclipse Adoptium JDK folder.

The Microsoft Store Hijack:

The Problem: When Spark tries to process data, it creates "worker" processes by calling the command python. Windows 11 intercepts this generic command and tries to open the Microsoft Store, causing the worker to time out and crash (Python worker failed to connect back).

The Solution: We set the PYSPARK_PYTHON and PYSPARK_DRIVER_PYTHON variables to point directly to your actual python.exe file, bypassing Windows entirely.

### Why Cassandra for Big Data
Cassandra works well for large data systems because it stays fast and available under heavy load.

#### 1) High write performance
Cassandra handles very fast writes, even with large data streams.
Good for: logs, analytics events, time-series data.

#### 2) Tunable consistency (AP system)
You can choose consistency level based on your need:
- Strong consistency: more accurate, higher latency
- Eventual consistency: faster, data syncs shortly after

#### 3) Horizontal scalability
Cassandra scales by adding more machines (nodes).
It is designed for multi-node clusters from the start.

#### 4) Fast data distribution
Data placement uses partition keys and hashing.
This avoids central routing and distributes load efficiently.
